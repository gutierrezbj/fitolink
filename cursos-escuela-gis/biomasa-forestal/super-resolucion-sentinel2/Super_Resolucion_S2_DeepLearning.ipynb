El módulo S2DR3 obtendrá datos de Sentinel-2 para la ubicación y los datos proporcionados y superresolverá las 10 bandas multiespectrales desde la resolución original de 10 m y 20 m hasta la resolución espacial objetivo de 1 m/px. La salida es una imagen TIF georreferenciada multiespectral de 10 bandas y 1 m/px. El producto de salida se generará en la ruta del sistema de archivos local '/content/output', que contendrá 4 productos:

* S2L2Ax10_T[MGRS]-[FECHA]-[UID]_MS.tif – Imagen multiespectal de 10 bandas
* S2L2Ax10_T[MGRS]-[FECHA]-[UID]_TCI.tif – Imagen RGB en color verdadero
* S2L2Ax10_T[MGRS]-[FECHA]-[UID]_NDVI.tif – Imagen NDVI en pseudocolor
* S2L2Ax10_T[MGRS]-[FECHA]-[UID]_IRP.tif – Imagen preudocolor de Ifra-rojo

El orden de las 10 bandas del producto multiespectral es:
B02, B03, B04, B05, B06, B07, B08, B8A, B11, B12.
